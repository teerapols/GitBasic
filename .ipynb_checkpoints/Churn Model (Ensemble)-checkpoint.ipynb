```python
import pandas as pd
import numpy as np

# Data Viz library
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
Churn = pd.read_excel('Churn_Modeling.xlsx', sheet_name = 'Train')

# ดูข้อมูล head 5 บรรทัดแรก

Churn.head()
```

```python
Churn
```

```python
Churn.replace([np.inf, -np.inf], np.nan)

Churn.isnull().any() # function check ค่าว่าง ถ้ามีค่าว่าง return true
```

```python
X = Churn.drop('Exited',axis=1)
y = Churn['Exited']
```

```python
X.head()
```

```python
X
```

```python
y.head()
```

```python
X = pd.get_dummies(X, columns=['Surname', 'Geography','Gender'])
X.head()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
# random_state คือ ชุดเลขสุ่ม หากต้องการผลเหมือนกัน ให้ระบุ 101 
# การกำหนด ratio = 1.0 คือให้ สุ่มเพิ่มจำนวนในแต่ละคลาสให้เท่ากัน

SM = SMOTE(random_state=101,sampling_strategy='auto')

X_res, y_res = SM.fit_sample(X, y)
```

```python
# แสดงจำนวนก่อนสุ่ม OverSampling

print("Before OverSampling, counts 'leave': {}".format(sum(y== "leave")))
print("Before OverSampling, counts 'stay': {} ".format(sum(y== "stay")))
```

```python
# แสดงจำนวนหลังสุ่ม OverSampling

print("After OverSampling, counts 'leave': {}".format(sum(y_res == "leave")))
print("After OverSampling, counts 'stay': {} ".format(sum(y_res == "stay")))
```

```python
# นำเข้า train test split function

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.30, random_state=55)
```

```python
# Voting Ensemble for Classification
import pandas
from sklearn import model_selection
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier


# create the sub models
model1 = DecisionTreeClassifier()
model2 = KNeighborsClassifier(n_neighbors = 1)
model3 = LogisticRegression()
model4 = RandomForestClassifier(max_features = 11, random_state = 55)

model = VotingClassifier(estimators=[('DT', model1), ('Knn', model2)
                                    , ('LG', model3),('RF',model4)], voting='hard')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
# ทำนายค่ากับชุด X_test

pred = model.predict(X_test)
```

```python
# นำเข้า classification report และ confusion matrix

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# สั่งพิมพ์รายงาน ค่าการวัดผล และประสิทธิภาพการทำนายของตัวแบบ และค่าใน confusion matrix

print(classification_report(y_test, pred))
print('\n')
print(confusion_matrix(y_test, pred))
```

```python
def find_error_classification(y_test, pred):
    N = y_test.shape[0]
    error = (100/N)*(y_test!= pred).sum()
    return error
```

```python
error_Test =find_error_classification(y_test, pred)
```
