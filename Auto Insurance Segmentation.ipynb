## Outlier Detection

```python
Q1_Claim = feature06.TotalClaimAmount.quantile(0.25)
Q3_Claim = feature06.TotalClaimAmount.quantile(0.75)
IQR = Q3_Claim - Q1_Claim

Lower = Q1_Claim - 1.5*IQR
Upper = Q3_Claim + 1.5*IQR

print (Q1_Claim , Q3_Claim, IQR, Lower, Upper)
```

```python
# Outlier Total Claim

feature07=feature06[(feature06.TotalClaimAmount>Lower)&(feature06.TotalClaimAmount<Upper)]
feature07.head()
```

```python
feature07.shape
```

```python
feature07.describe()
```

```python
Q1_Income = feature07.Income.quantile(0.25)
Q3_Income = feature07.Income.quantile(0.75)
IQR2 = Q3_Income - Q1_Income

Lower2 = Q1_Income - 1.5*IQR2
Upper2 = Q3_Income + 1.5*IQR2

print (Q1_Income , Q3_Income, IQR2, Lower2, Upper2)
```

```python
# Outlier Income

feature08=feature07[(feature07.Income>Lower2)&(feature07.Income<Upper2)]
feature08.head()
```

```python
feature08.shape
```

```python
feature08.describe()
```

## PCA

```python
# Train test split

from sklearn.model_selection import train_test_split
X_train,X_test =train_test_split(feature08,test_size = 0.05,random_state = 43232)
print(X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


scaler = StandardScaler()
X_train_std = pd.DataFrame(scaler.fit_transform(X_train),columns=X_train.columns,index=X_train.index)
X_train_std.head()
```

```python
pca = PCA(.90)
X_train_pca = pd.DataFrame(pca.fit_transform(X_train_std),index=X_train.index)
X_train_pca.head()
```

```python
X_train_pca.shape
```

## K-Mean

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model,k=(1,10))

visualizer.fit(X_train_pca)
visualizer.show()
```

```python
kmean_pca = KMeans(n_clusters=5,init = 'k-means++')
kmean_pca.fit(X_train_pca)
```

```python
predicted_pca =pd.DataFrame(kmean_pca.predict(X_train_pca),columns=["Cluster"],index=X_train_pca.index)
predicted_pca.head()
```

```python
#Silloutte score
from sklearn.metrics import silhouette_score
print(silhouette_score(X_train_pca,predicted_pca))
```

```python
print(kmean_pca.score(X_train_pca))
```

```python
from yellowbrick.cluster import SilhouetteVisualizer

model = KMeans(5, random_state=42)
visualizer = SilhouetteVisualizer(model, colors='yellowbrick')

visualizer.fit(X_train_pca)        # Fit the data to the visualizer
visualizer.show()    
```
