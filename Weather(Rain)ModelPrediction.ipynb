## Model Random Forest

## แบ่ง Train Test Split

```python
# นำเข้า train test split function

from sklearn.model_selection import train_test_split
```

```python
# ใช้วิธี Hold out แบ่ง Test 20 และ Train 70

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101,stratify = y)
```

## สร้าง Model

```python
from sklearn.ensemble import RandomForestClassifier# ตั้งชื่อตัวแบบว่า "rf" ให้สุ่ม 500 trees

rf = RandomForestClassifier(n_estimators = 500)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_

#Index(['MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation', 'Sunshine',
      # 'WindGustSpeed', 'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am',
      # 'Humidity3pm', 'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm',
      # 'Temp9am', 'Temp3pm'],
      # dtype='object')
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
# สั่งพิมพ์รายงาน ค่าการวัดผล และประสิทธิภาพการทำนายของตัวแบบ

print(classification_report(y_test, y_pred))
```

```python
Prediction = pd.DataFrame(y_pred, columns = ["Prediction"], index = X_test.index)
Prediction.head()
```

```python
y_test
```

```python
result = pd.concat([y_test,Prediction], axis = 1)
result
```
