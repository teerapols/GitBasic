## Model Random Forest

## แบ่ง Train Test Split

```python
# นำเข้า train test split function

from sklearn.model_selection import train_test_split
```

```python
# ใช้วิธี Hold out แบ่ง Test 20 และ Train 70

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101,stratify = y)
```

## สร้าง Model

```python
from sklearn.ensemble import RandomForestClassifier# ตั้งชื่อตัวแบบว่า "rf" ให้สุ่ม 500 trees

rf = RandomForestClassifier(n_estimators = 500)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
# สั่งพิมพ์รายงาน ค่าการวัดผล และประสิทธิภาพการทำนายของตัวแบบ

print(classification_report(y_test, y_pred))
```

```python
Prediction = pd.DataFrame(y_pred, columns = ["Prediction"], index = X_test.index)
Prediction.head()
```

```python
y_test
```

```python
result = pd.concat([y_test,Prediction], axis = 1)
result
```

## Adjust hyperparameter

```python
from sklearn.model_selection import GridSearchCV

# Create the parameter grid based on the results of random search 
param_grid = {
    'max_features': [4,6,8],
    'min_samples_leaf': [1,3,5],
    'min_samples_split': [2,5,12],
    'n_estimators': [100,300,500]
}

# Create a based model
rf = RandomForestClassifier()

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, verbose = 3)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# ดู default parameter

grid_search.best_estimator_
```

## สร้าง Model RF ใหม่โดยใช้ Best estimator จาก Grid Search

```python
rf_fine =              RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=None, max_features=8,
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=500,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False)
```

##  Train Model ใหม่ด้วย RF

```python
# สร้าง Model "rf_fine"

rf_fine.fit(X_train, y_train)
```

## Prediction Model ใหม่

```python
# ทำนายค่าในชุด test set ที่แบ่งไว้ 30%

Predictionfine= rf_fine.predict(X_test)
```

## Evaluation Model ใหม่

```python
# นำเข้า classification report และ confusion matrix

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# สั่งพิมพ์รายงาน ค่าการวัดผล และประสิทธิภาพการทำนายของตัวแบบ และค่าใน confusion matrix

print(classification_report(y_test, Predictionfine))
print('\n')
print(confusion_matrix(y_test, Predictionfine))
```
