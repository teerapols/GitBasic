## Model Random Forest

## à¹à¸šà¹ˆà¸‡ Train Test Split

```python
# à¸™à¸³à¹€à¸‚à¹‰à¸² train test split function

from sklearn.model_selection import train_test_split
```

```python
# à¹ƒà¸Šà¹‰à¸§à¸´à¸˜à¸µ Hold out à¹à¸šà¹ˆà¸‡ Test 20 à¹à¸¥à¸° Train 70

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101,stratify = y)
```

## à¸ªà¸£à¹‰à¸²à¸‡ Model

```python
from sklearn.ensemble import RandomForestClassifier# à¸•à¸±à¹‰à¸‡à¸Šà¸·à¹ˆà¸­à¸•à¸±à¸§à¹à¸šà¸šà¸§à¹ˆà¸² "rf" à¹ƒà¸«à¹‰à¸ªà¸¸à¹ˆà¸¡ 500 trees

rf = RandomForestClassifier(n_estimators = 500)
```

```python
rf.fit(X_train, y_train)
```

```python
rf.feature_importances_
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
# à¸ªà¸±à¹ˆà¸‡à¸žà¸´à¸¡à¸žà¹Œà¸£à¸²à¸¢à¸‡à¸²à¸™ à¸„à¹ˆà¸²à¸à¸²à¸£à¸§à¸±à¸”à¸œà¸¥ à¹à¸¥à¸°à¸›à¸£à¸°à¸ªà¸´à¸—à¸˜à¸´à¸ à¸²à¸žà¸à¸²à¸£à¸—à¸³à¸™à¸²à¸¢à¸‚à¸­à¸‡à¸•à¸±à¸§à¹à¸šà¸š

print(classification_report(y_test, y_pred))
```

```python
Prediction = pd.DataFrame(y_pred, columns = ["Prediction"], index = X_test.index)
Prediction.head()
```

```python
y_test
```

```python
result = pd.concat([y_test,Prediction], axis = 1)
result
```

## Adjust hyperparameter

```python
from sklearn.model_selection import GridSearchCV

# Create the parameter grid based on the results of random search 
param_grid = {
    'max_features': [4,6,8],
    'min_samples_leaf': [1,3,5],
    'min_samples_split': [2,5,12],
    'n_estimators': [100,300,500]
}

# Create a based model
rf = RandomForestClassifier()

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, verbose = 3)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# à¸”à¸¹ default parameter

grid_search.best_estimator_
```

## à¸ªà¸£à¹‰à¸²à¸‡ Model RF à¹ƒà¸«à¸¡à¹ˆà¹‚à¸”à¸¢à¹ƒà¸Šà¹‰ Best estimator à¸ˆà¸²à¸ Grid Search

```python
rf_fine =              RandomForestClassifier(bootstrap=True, ccp_alpha=0.0, class_weight=None,
                       criterion='gini', max_depth=None, max_features=8,
                       max_leaf_nodes=None, max_samples=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=500,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False)
```

##  Train Model à¹ƒà¸«à¸¡à¹ˆà¸”à¹‰à¸§à¸¢ RF

```python
# à¸ªà¸£à¹‰à¸²à¸‡ Model "rf_fine"

rf_fine.fit(X_train, y_train)
```

## Prediction Model à¹ƒà¸«à¸¡à¹ˆ

```python
# à¸—à¸³à¸™à¸²à¸¢à¸„à¹ˆà¸²à¹ƒà¸™à¸Šà¸¸à¸” test set à¸—à¸µà¹ˆà¹à¸šà¹ˆà¸‡à¹„à¸§à¹‰ 30%

Predictionfine= rf_fine.predict(X_test)
```

## Evaluation Model à¹ƒà¸«à¸¡à¹ˆ

```python
# à¸™à¸³à¹€à¸‚à¹‰à¸² classification report à¹à¸¥à¸° confusion matrix

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# à¸ªà¸±à¹ˆà¸‡à¸žà¸´à¸¡à¸žà¹Œà¸£à¸²à¸¢à¸‡à¸²à¸™ à¸„à¹ˆà¸²à¸à¸²à¸£à¸§à¸±à¸”à¸œà¸¥ à¹à¸¥à¸°à¸›à¸£à¸°à¸ªà¸´à¸—à¸˜à¸´à¸ à¸²à¸žà¸à¸²à¸£à¸—à¸³à¸™à¸²à¸¢à¸‚à¸­à¸‡à¸•à¸±à¸§à¹à¸šà¸š à¹à¸¥à¸°à¸„à¹ˆà¸²à¹ƒà¸™ confusion matrix

print(classification_report(y_test, Predictionfine))
print('\n')
print(confusion_matrix(y_test, Predictionfine))
```
