## 1. Import Library

```python
import pandas as pd
import numpy as np

# Data Viz library
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## 2.Read Data

```python
Loan = pd.read_excel('loanprediction.xlsx', sheet_name = 'Train')

# ดูข้อมูล head 5 บรรทัดแรก

Loan.head()
```

## 3. Data Preparation (Clean & Fill in missing value)

```python
Loan.replace([np.inf, -np.inf], np.nan)

Loan.isnull().any() # function check ค่าว่าง ถ้ามีค่าว่าง return true
```

## แบ่งชุด X กับ Y , Train กับ Test 

```python
X = Loan.drop('loan_condition',axis=1)
y = Loan['loan_condition']
```

```python
X.head()
```

```python
y.head()
```

```python
X = pd.get_dummies(X, columns=['purpose', 'verification_status','grade','sub_grade','emp_length','home_ownership','addr_state','term'])
X.head()
```

```python
X
```

```python
y.head()
```

## Import SMOTE sampling แก้ปัญหา Imbalance

```python
from imblearn.over_sampling import SMOTE
```

```python
# random_state คือ ชุดเลขสุ่ม หากต้องการผลเหมือนกัน ให้ระบุ 101 
# การกำหนด ratio = 1.0 คือให้ สุ่มเพิ่มจำนวนในแต่ละคลาสให้เท่ากัน

SM = SMOTE(random_state=101,sampling_strategy='auto')

X_res, y_res = SM.fit_sample(X, y)
```

```python
# แสดงจำนวนก่อนสุ่ม OverSampling

print("Before OverSampling, counts 'Good': {}".format(sum(y== "Good")))
print("Before OverSampling, counts 'Bad': {} ".format(sum(y== "Bad")))
```

```python
# แสดงจำนวนหลังสุ่ม OverSampling

print("After OverSampling, counts 'Good': {}".format(sum(y_res == "Good")))
print("After OverSampling, counts 'Bad': {} ".format(sum(y_res == "Bad")))
```

## แบ่งชุด X กับ Y , Train 70% และ Test 30%

```python
# นำเข้า train test split function

from sklearn.model_selection import train_test_split
```

```python
# ใช้วิธี Hold out แบ่ง Test 30 และ Train 70

X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.30, random_state=55)
```

## 4.สร้าง Model Random Forest

```python
from sklearn.ensemble import RandomForestClassifier# ตั้งชื่อตัวแบบว่า "rf" ให้สุ่ม 500 trees

rf = RandomForestClassifier(n_estimators = 500)
```

## 5.Train Model ด้วย Random Forest

```python
rf.fit(X_train, y_train)
```

## 6. Predict ทำนายผล

```python
# ทำนายค่าในชุด test set ที่แบ่งไว้ 30%

pre_rf = rf.predict(X_test)
```

```python
pre_rf
```

## 7.Evaluate Model

```python
# นำเข้า classification report และ confusion matrix

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# สั่งพิมพ์รายงาน ค่าการวัดผล และประสิทธิภาพการทำนายของตัวแบบ และค่าใน confusion matrix

print(classification_report(y_test, pre_rf))
print('\n')
print(confusion_matrix(y_test, pre_rf))
```

```python
def find_error_classification(y_test, pre_rf):
    N = y_test.shape[0]
    error = (100/N)*(y_test!= pre_rf).sum()
    return error
```

```python
error_Test =find_error_classification(y_test, pre_rf)
```

```python
error_Test
```

```python
# แปลงค่า array ของผล pred ให้เป็น DataFrame
# ตั้งชื่อ columns ว่า 'Predict'

columns_new = ['Predict']

# pass in array and columns
prediction = pd.DataFrame(pre_rf, columns=columns_new)
prediction.head()
```

```python
y_test
```

```python
# แปลงค่า array ของผล pred ให้เป็น DataFrame
# ตั้งชื่อ columns ว่า 'Predict'

columns_new = ['Yhat']

# pass in array and columns
Yhat = pd.DataFrame(y_test, columns=columns_new)
Yhat.head()
```

```python
result = pd.concat([Yhat,prediction], axis = 1)
result
```

## 8.Import Prediction Result to Excel File

```python
result.to_excel('Loan_Predict.xlsx',sheet_name='Sheet1',index=False)
```

## 9. ปรับแต่ง Hyper Parameter เพื่อเพิ่มประสิทธิภาพ Model โดยใช้ Grid search

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV

# Create the parameter grid based on the results of random search 
param_grid = {
    'max_features': [4,6,8],
    'min_samples_leaf': [1,3,5],
    'min_samples_split': [2,5,12],
    'n_estimators': [100,300,500]
}

# Create a based model
rf = RandomForestClassifier()

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, verbose = 3)
```

```python
# Fit the grid search to the data
grid_search.fit(X_train, y_train)
```

```python
# ดู default parameter

grid_search.best_estimator_
```

## 10. สร้าง Model RF ใหม่โดยใช้ Best estimator จาก Grid Search

```python
rf_fine   =        RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',
                       max_depth=None, max_features=4, max_leaf_nodes=None,
                       min_impurity_decrease=0.0, min_impurity_split=None,
                       min_samples_leaf=1, min_samples_split=2,
                       min_weight_fraction_leaf=0.0, n_estimators=300,
                       n_jobs=None, oob_score=False, random_state=None,
                       verbose=0, warm_start=False)
```

## 11. Train Model ใหม่ด้วย RF

```python
# สร้าง Model "rf_fine"

rf_fine.fit(X_train, y_train)
```

## 12. Prediction Model ใหม่

```python
# ทำนายค่าในชุด test set ที่แบ่งไว้ 30%

predictionfine = rf_fine.predict(X_test)
```

## 13. Evaluation Model ใหม่

```python
# นำเข้า classification report และ confusion matrix

from sklearn.metrics import classification_report, confusion_matrix
```

```python
# สั่งพิมพ์รายงาน ค่าการวัดผล และประสิทธิภาพการทำนายของตัวแบบ และค่าใน confusion matrix

print(classification_report(y_test, predictionfine))
print('\n')
print(confusion_matrix(y_test, predictionfine))
```
