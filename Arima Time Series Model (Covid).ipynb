```python
# แปลง date ให้อยู่ใน Format วันที่ (จาก object เป็น datetime)
t_2['Date'] = t_2['Date'].astype('datetime64')
t_2.dtypes
```

```python
# Grouping Data by Date และแยกเฉพาะผู้ติดเชิ้อ
Covid = t_2.groupby('Date')[['Confirmed']].sum().round(2)
Covid.head()
```

## Decompose

```python
import statsmodels.api as sm
from pylab import rcParams

rcParams['figure.figsize'] = 11, 9  # aspect ratio

decomposition = sm.tsa.seasonal_decompose(Covid, model='additive')
fig = decomposition.plot()
plt.show()
```

## ADF Test, Stationary checking

```python
#ADF Test, Stationary checking
from statsmodels.tsa.stattools import adfuller, kpss

# ADF Test
result = adfuller(Covid.Confirmed.values, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critical Values:')
    print(f'   {key}, {value}')

# KPSS Test
result = kpss(Covid.Confirmed.values, regression='c')
print('\nKPSS Statistic: %f' % result[0])
print('p-value: %f' % result[1])
for key, value in result[3].items():
    print('Critical Values:')
    print(f'   {key}, {value}')
```

## Plot ACF,PACF

```python
# ACF, PACF
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf



# Calculate ACF and PACF upto 50 lags
# acf_50 = acf(df.value, nlags=50)
# pacf_50 = pacf(df.value, nlags=50)

# Draw Plot
fig, axes = plt.subplots(1,2,figsize=(16,3), dpi= 100)
plot_acf(Covid.Confirmed.tolist(), lags=50, ax=axes[0])
plot_pacf(Covid.Confirmed.tolist(), lags=50, ax=axes[1])
```

```python
# ACF, PACF
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.figsize':(9,7), 'figure.dpi':120})

# Original Series
fig,axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(Covid1.Confirmed); axes[0, 0].set_title('Original Series')
plot_acf(Covid1.Confirmed,ax=axes[0, 1])


# 1st Differencing
axes[1, 0].plot(Covid1.Confirmed.diff()); axes[1, 0].set_title('1st Order Differencing')
plot_acf(Covid1.Confirmed.diff().dropna(), ax=axes[1, 1])

# 2nd Differencing
axes[2, 0].plot(Covid1.Confirmed.diff().diff()); axes[2, 0].set_title('2nd Order Differencing')
plot_acf(Covid1.Confirmed.diff().diff().dropna(), ax=axes[2, 1])

result = adfuller(Covid1.Confirmed.diff().diff().dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

plt.show()
```

```python
# PACF plot of 2nd order differenced series
plt.rcParams.update({'figure.figsize':(9,3), 'figure.dpi':120})

fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(Covid1.Confirmed.diff().diff()); axes[0].set_title('2nd Differencing')
axes[1].set(xlim=(0, 50), ylim=(0,1.2))
plot_pacf(Covid1.Confirmed.diff().diff().dropna(), ax=axes[1])

plt.show()
```

## แบ่ง Train and Test Split

```python
len(Covid)
```

```python
train_len = len(Covid)-31
```

```python
train = Covid.iloc[:train_len]
test = Covid.iloc[train_len:]
```

```python
len(train)
```

```python
len(test)
```

## Time Series Machine Learning Model Arima

```python
# auto arima
import pmdarima as pm
from pmdarima.arima import auto_arima
```

```python
# 2,2,2 ARIMA Model
#(p, d, q)
# p lags for AR, using PACF
#d is order of differencing
#q window for MA, using ACF

# fit stepwise auto-ARIMA
stepwise_model = pm.auto_arima(train, start_p=2, start_q=2,
                             max_p=3, max_q=3, m=3, 
                             start_P=2, seasonal=True,
                             d=2, D=2, trace=True,
                             error_action='ignore',  # don't want to know if an order does not work
                             suppress_warnings=True,  # don't want convergence warnings
                             stepwise=True)  # set to stepwise
```

```python
stepwise_model.summary()
```

## เรียนรู้

```python
stepwise_model.fit(train)
```

## Predict with Test Set

```python
# ระยะเวลาการ Test ตั้งแต่ 2020-04-05 ถึง 2020-04-11 เป็น 7 วัน นำไปใส่เป็นจำนวน n_periods= 7

future_forecast = stepwise_model.predict(n_periods=31)
# This returns an array of predictions:
print(future_forecast)
```

```python
future_forecast = pd.DataFrame(future_forecast,index = test.index,columns=['Prediction'])

pred = pd.concat([test,future_forecast],axis=1)
pred.round(2)
```

## Evaluate Model

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
MAE = (mean_absolute_error(pred['Confirmed'],pred.Prediction)).round(2)
MSE = (mean_squared_error(pred['Confirmed'],pred.Prediction)).round(2)
RMSE =(np.sqrt(mean_squared_error(pred['Confirmed'],pred.Prediction))).round(2)
```

```python
print("MAE = ", MAE), 
print("MSE = ", MSE),
print("RMSE= ", RMSE)
```
