```python
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
import pandas as pd

from tensorflow.keras.layers import Dense, Input, Conv2D, Flatten,GlobalAveragePooling2D, Dropout #GlobalAveragePooling4D#MaxPooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator #แปลง
from tensorflow.keras.optimizers import Adam
from keras.preprocessing import image
```

```python
batch_size = 128
num_classes = 2
target_size = 224
#target_size2 = 111
#target_size = 299
learning_rate = 0.001
dense_size = 1024
#dropout_rate = 0.05
```

```python
datagen_train = ImageDataGenerator(rescale=1./255)
```

```python
datagen_test = ImageDataGenerator(rescale=1./255)           
```

```python
train_datagen = datagen_train.flow_from_directory('C:/Users/Teerapol/Image Analytic/chest_xray/train',
  target_size=(target_size,target_size),
  batch_size=batch_size
)
```

```python
test_datagen = datagen_test.flow_from_directory('C:/Users/Teerapol/Image Analytic/chest_xray/test',
  target_size=(target_size,target_size),
  batch_size=batch_size
)
```

```python
x_batch, y_batch = next(train_datagen)
print(x_batch.shape)

fig = plt.figure(figsize=(16,16))

#show only the first 64 samples if batch size is over 64
for i in range(32):
  fig.add_subplot(8, 8, 1 + i, xticks=[], yticks=[])

  #mobileNet preprocessing convert color range to [-1,1]
  #need to change to [0,1] to show color properly
  plt.imshow((x_batch[i]+1)/2) #คำสั่งแก้ range ให้เป็นช่วง [0,1],[0,255]
plt.show()
```

```python
x_batch, y_batch = next(test_datagen)

print(x_batch.shape)

fig = plt.figure(figsize=(16,16))

#show only the first 64 samples if batch size is over 64
for i in range(32):
  fig.add_subplot(8, 8, 1 + i, xticks=[], yticks=[])
  #mobileNet preprocessing convert color range to [-1,1]
  #need to change to [0,1] to show color properly
  plt.imshow((x_batch[i]+1)/2)
plt.show()
```

```python
tensorboard = keras.callbacks.TensorBoard(log_dir='./logs')
```

```python
input_img = Input(shape=(target_size,target_size,3),name='input')
x = Conv2D(32, (3, 3), padding='same', activation='relu')(input_img)
x = GlobalAveragePooling2D()(x)
#x = MaxPooling2D((2,2))(x)
#x = Input(shape=(target_size2,target_size2,32)(x)
#x = Input(shape=(target_size2,target_size2,32),name='input')
#x = Conv2D(32, (3, 3), padding='same', activation='relu')(x)
#x = GlobalAveragePooling2D()(x)
#x = MaxPooling2D((2,2))(x)
x = Flatten()(x)
x = Dense(dense_size, activation='relu')(x)
x = Dropout(0.5)(x)
x = Dense(dense_size, activation='relu')(x)
x = Dropout(0.5)(x)
out = Dense(num_classes, activation='sigmoid')(x)

model = Model(input_img, out)
#model.compile(optimizer=Adam(lr=learning_rate),
model.compile(optimizer=Adam(lr=learning_rate),              
                        loss='categorical_crossentropy',
                       metrics=['accuracy'])

model.summary()
```

```python
model.fit_generator(
  train_datagen,
  steps_per_epoch=5216//batch_size,
  shuffle=True, #เรียงภาพแบบสุ่ม ไม่เรียงลำดับ
  epochs=10,
)
```

```python
model.evaluate_generator(test_datagen,verbose=1)
```

```python
# Predict
result = model.predict(test_datagen)
print(result)
print(np.round(result))
```
