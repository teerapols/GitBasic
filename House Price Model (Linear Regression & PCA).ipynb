## ทำ PCA ต้องทำ Standardization ก่อน

```python
# เริ่มด้วยการนำเข้า StandardScaler เป็นตัวปรับ Normalize scale

from sklearn.preprocessing import StandardScaler
```

```python
# ตั้งชื่อ function "stdz"

stdz = StandardScaler()
```

```python
# สร้าง function "stdz"

stdz.fit(X)# สั่งให้ทำ Normalize ตัวแปร Independent ทุกตัว

```

```python
# สั่งให้ทำ Normalize ตัวแปร Independent ทุกตัว

scaled_Ind = stdz.transform(X)
```

```python
scaled_Ind
```

```python
# เตรียมสร้าง ชื่อ column ของชุดข้อมูล salary

X.columns
```

```python
# สร้างชุดข้อมูลที่ทำ standardize แล้ว จะมี  center of mean = 0  เราตั้งชื่อว่า "salary_std" 

X_std = pd.DataFrame(scaled_Ind, columns = X.columns)
```

```python
X_std.head(3)
```

## Import PCA ลด Feature, ลด Dimension

```python
from sklearn.decomposition import PCA

for i in range(11):
    pca = PCA(n_components=i)
    pca.fit(X_std)

    print(sum(pca.explained_variance_ratio_))

```

```python
pca = PCA(.90)
X_pca = pd.DataFrame(pca.fit_transform(X_std), index =X_std.index)

print (X_pca .shape)
```

```python
X_std.shape
```

```python
pca = PCA(.60)
X_pca = pd.DataFrame(pca.fit_transform(X_std), index =X_std.index)

print (X_pca .shape)
```

```python
X_pca.head()
```

```python
# เริ่มเข้าสู่ขั้นตอนของการ Train model
# นำเข้า train test split function

from sklearn.model_selection import train_test_split
```

```python
# X คือ attributes, y คือ target class

# ใช้วิธี Hold out แบ่ง Test 30 และ Train 70

X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.20, random_state=55)
```

```python
# Import "LinearRegression" from scikit-learn

from sklearn.linear_model import LinearRegression
```

```python
# ตั้งชื่อ Linear Regression model ว่า "lm"

lm = LinearRegression()
```

```python
# สร้าง model regression

lm.fit(X_train, y_train)
```

```python
pred = lm.predict(X_test)

pred
```

```python
y_test.head()
```

```python
y_test.reset_index(drop=True, inplace=True)

y_test
```

```python
plt.scatter(y_test,pred)
```

```python
# Residual are the difference between the actual values and predicted values

# Residual --> Normal distribution means your model was a correct prediction for the data
sns.distplot((y_test-pred),bins=100);
```

```python
from sklearn import metrics
```

```python
print('MAE', metrics.mean_absolute_error(y_test, pred))
print('MSE', metrics.mean_squared_error(y_test, pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```

```python
metrics.explained_variance_score(y_test, pred)
```
